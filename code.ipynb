```python
from IPython import display
from base64 import b64decode
from PIL import Image
import io
import numpy as np
import PIL.ImageOps    


print("0.")
base64_data = ""   
binary_data = b64decode(base64_data)
print("1.")
# display.Image(binary_data)
print(len(binary_data))


print("2.")
image = Image.open(io.BytesIO(binary_data))
inverted_image = PIL.ImageOps.invert(image)
print(image.format)

img_byte_arr = io.BytesIO()
inverted_image.save(img_byte_arr, format=image.format)
new_binary_data = img_byte_arr.getvalue()
print(len(new_binary_data))
   

print("3.")
# image.show()
display.Image(new_binary_data)
```
